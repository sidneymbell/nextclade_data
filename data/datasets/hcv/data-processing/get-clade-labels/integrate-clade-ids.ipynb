```python
import pandas as pd
```

```python
original_meta = pd.read_csv('./data/hcv-meta.tsv', sep='\t', index_col='strain')
inferred_meta = pd.read_csv('/Users/sbell/Downloads/nextstrain__metadata (2).tsv', sep='\t', index_col='strain')
```

```python
del original_meta['clade_membership']
inferred_meta['clade_membership'] = inferred_meta['clade_membership'].map(lambda i: round(i))
```

```python
original_meta = original_meta.join(inferred_meta['clade_membership'], on='strain')
```

```python
original_meta.reset_index(inplace=True)
```

```python
original_meta.to_csv('./data/hcv-meta.tsv', sep='\t')
```
